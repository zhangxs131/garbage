#6. 词干化
给定的文本中，将每个token转换为它的词根形式.使用nltk.stem中的PorterStemmer


```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_tokens=[]

text='''“哇，这个新来的boy真的super handsome!”,
 “这个timetable做的not very good”,
 “coffee 我们需要meet一下然后看看tomorrow怎么安排”,'''

for token in nltk.word_tokenize(text):
  stemmed_tokens.append(stemmer.stem(token))

print((' ').join(stemmed_tokens))
```

#9. 度量余弦相似度
使用sklearn中的tf-idf算法提取文本特征，然后通过计算特征余弦来评估文本相似度。

```python
import jieba
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

text1 = '涉黄涉政检测：对文本内容做涉黄涉政检测，满足相应政策要求'
text2 = '垃圾评论过滤：在论坛发言或用户评论中，过滤文本中的垃圾广告，提升文本总体质量'
text3 = '情感分析：对用户评论等文本内容做情感分析，指导决策与运营'
text4 = '自动标签提取：自动提取文本重要内容生成关键性标签，在此基础之上拓展更多功能形式'
text5 = '文本自动分类：通过对文本内容进行分析，给出文本所属的类别和置信度，支持二级分类。'

documents=[text1,text2,text3,text4,text5]

vectorizer=CountVectorizer(
    stop_words=my_stopwords,
    tokenizer=lambda x : ' '.join(jieba.lcut(x))
)
tf_matrix=vectorizer.fit_transform(documents)

doc_term_matrix=tf_matrix.todense()

df=pd.DataFrame(doc_term_matrix)

print(cosine_similarity(df,df))
```

```python
from  sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

vectorizer=TfidfVectorizer(
    stop_words=my_stopwords,
    max_features=1000,
    max_df=0.4,
    smooth_idf=True
)

nmf_matrix=vectorizer.fit_transform(sentence_pro)

nmf_model=NMF(n_components=5)
nmf_model.fit(nmf_matrix)

def print_topics_nmf(model,vectorizer,top_n=5):
  for idx,topic in enumerate(model.components_):
    print('Topic %d:{}'.format(idx+1))
    print(
        [(vectorizer.get_feature_names()[i],topic[i]) for i in topic.argsort()[:top_n-1:-1]]
    )
print_topics_nmf(nmf_model,vectorizer)
```

#15. 使用gensim中tfidf模型来获取句向量表示

```python
import numpy as np
from gensim import corpora
from gensim.utils import simple_preprocess

dictionary = corpora.Dictionary()
#从文档中创建词袋模型

doc_tokenized=[[w.strip() for w in jieba.lcut(s)] for s in sentences]

Bow_corpus=[dictionary.doc2bow(doc,allow_update=True) for doc in doc_tokenized]

tfidf=gensim.models.TfidfModel(Bow_corpus)

#获取其中一个语句的TF-IDF句向量表示
print(tfidf[Bow_corpus[4]])
```

#16. 创建bigrams和trigrams

*    可以使用nltk 中ngrams方法抽取
*    可以使用gensim中Phraser 来抽取


```python
from gensim.models import Phrases
from gensim.models.phrases import Phraser

#创建一个bigram phraser
bigram=Phrases(doc_tokenized,min_count=1,threshold=8,delimiter=b'_')
bigram_pharser=Phraser(bigram)

for sent in doc_tokenized:
  tokens_=bigram_pharser[sent]
  print([i.strip() for i in tokens_ if i.find('_')>0])
```

#18. 使用GPT2预训练模型自动生成诗歌

transformers中pipeline

```python
from transformers import GPT2LMHeadModel,BertTokenizerFast,pipeline

tokenizer=BertTokenizerFast.from_pretrained('gaochangkuan/model_dir')
model = GPT2LMHeadModel.from_pretrained('gaochangkuan/model_dir', pad_token_id=tokenizer.eos_token_id)

nlp=pipeline('text-generation',model=model,tokenizer=tokenizer)

result=nlp('北国冬日')
print(result)
```
